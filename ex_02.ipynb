```python
import pymongo
from random import randint
from faker import Faker
```

```python
conexao = pymongo.MongoClient('mongodb://localhost:27017/')
```

```python
# QUESTÃO 01

db = conexao["site"]
```

```python
# QUESTÃO 02
#Criando as coleções.
colecao_usuarios = db["usuarios"]
colecao_acessos = db["acessos"]

colecoes = db.list_collection_names()

print(colecoes)
```

```python
# QUESTÃO 03
# Atribuindo dados as coleções. Utilizado a biblioteca Faker para criar os dados aleatórios.
locales = 'pt-Br'
fake = Faker(locales)


id=1

for i in range(30):
    colecao_usuarios.insert_one(
        { 
    "_id": id,
    "nome": fake.name(),
    "idade": (randint(18,65)),
    "estado": fake.administrative_unit()
        }
    )
    id += 1
```

```python

id=1

for i in range(200):
    colecao_acessos.insert_one(
        { 
    "_id": id,
    "cpf": fake.ssn(),
    "data_nascimento": fake.date(),
    "e-mail": fake.email(),
    "profissão": fake.job(),
    "ano_matricula": fake.year()    
        }
    )
    id += 1

```

```python
# QUESTÃO 04
# A maiorias das datas está  cadastrada como anterior a 2021. Utilizando o $gte retorna 10 registros, destes sendo 5 igual a 2021.
x = colecao_acessos.find(
    {
        "ano_matricula": {'$lt': '2021'}
    }
)

for doc in x:
    print(doc)
```

```python
#Questão 5
#Como foi utilizado uma biblioteca para preencher os nomes, não nomes sem preenchimento.

x = colecao_usuarios.find(
    {
    'nome': {'$exists': False}
    }
)

for doc in x:
    print(doc)
```
